```python
sns.histplot(data['Age'],kde=True)
```

```python
data.columns
```

```python
sns.histplot(data['Annual_Premium'], kde=True, color='purple')
plt.title('count vs Annual_Premium')
```

```python
sns.countplot(x='Gender', data=data, palette='pastel')
plt.title('Gender Distribution')
```

```python
sns.countplot(x='Vintage', data=data, palette='Blues')
plt.title('Distribution of Policy Types')
```

```python
sns.countplot(x='Vintage', hue='Previously_Insured', data=data, palette='Set1')
plt.title('Policy Type vs Insurance Claims')
```

```python
sns.countplot(x='Gender', hue='Previously_Insured', data=data)
plt.title('Previously_Insured vs Gender')
```

```python
sns.countplot(x='Region_Code', hue='Previously_Insured', data=data)
plt.title('Previously_Insured vs Region Code')
```

```python
sns.countplot(x='Policy_Sales_Channel', hue='Vehicle_Age', data=data)
plt.title('vechile Age vs Policy Channel')
```

```python
sns.boxplot(x='Previously_Insured', y='Annual_Premium', data=data)
plt.title('Previously_Insured vs Annual_Premium')
```

```python
sns.countplot(x='Previously_Insured', data=data, palette='Set2')
plt.title('Insurance Claim Count')
```

```python
numeric_data = data.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(8,6))
```

```python
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
```

```python
sns.countplot(x='Vehicle_Damage', hue='Previously_Insured', data=data, palette='coolwarm')
plt.title('Vehicle Damage vs Insurance Claim')
```

```python
sns.boxplot(x='Policy_Sales_Channel', y='Annual_Premium', data=data, palette='Pastel1')
plt.title('Claim Likelihood vs Number of Policies')
```
